```python
import pandas as pd
```

```python
df = pd.read_csv('glassdoor_jobs.csv')
```

```python
df.head()
```

# Salary Parsing

```python

df = df[df["Salary Estimate"]!="-1"]
```

```python
salary = df["Salary Estimate"].apply(lambda x:x.split('(')[0])
```

```python
salary
```

```python
remove_k_doll = salary.apply(lambda x: x.replace('K','').replace('$',''))
```

```python
df['hourly'] = df['Salary Estimate'].apply(lambda x:1 if 'per hour' in x.lower() else 0)
```

```python
df.head()
```

```python
df['employer_provided_salary'] = df['Salary Estimate'].apply(lambda x:1 if 'employer provided salary:' in x.lower() else 0)
```

```python
removed_hr_emply = remove_k_doll.apply(lambda x:x.lower().replace('per hour','').replace('employer provided salary:',''))
```

```python
df['min_salary'] = removed_hr_emply.apply(lambda x:int(x.split('-')[0]))
```

```python
df['min_salary'].dtypes
```

```python
df['max_salary'] = removed_hr_emply.apply(lambda x:int(x.split('-')[1]))
```

```python
df.head()
```

```python
df['max_salary'].dtypes, df['min_salary'].dtypes
```

```python
df['avg_salary'] =(df['max_salary']+df['min_salary'])/2
```

```python
df.head()
```

# Company Name only

```python
df.head()
```

```python
company_txt = df.apply(lambda x:x['Company Name'] if x['Rating']<0 else x['Company Name'][:-3],axis=1)
```

```python
df['company_txt'] = company_txt
```

```python
df.head()
```

```python
df['company_txt'] = df['company_txt'].apply(lambda x:x.replace("\n",""))
```

```python
df.head()
```

# state field

```python
df['job_state'] = df['Location'].apply(lambda x:x.split(',')[0])
```

```python
df.head()
```

```python
df['job_state'].value_counts()
```

```python
df['same_state'] = df.apply(lambda x:1 if x['Location']==x['Headquarters'] else 0,axis=1)
```

```python
df.head()
```

# Age of the company

```python
df['age'] = df['Founded'].apply(lambda x:x if x<1 else 2020-x)
```

```python
df.head()
```

```python
df['Job Description'][0]
```

```python
#python 
df['python'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
```

```python
df['R'] = df['Job Description'].apply(lambda x: 1 if 'r_studio' in x.lower() else 0)
```

```python
df['spark'] = df['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() else 0)
```

```python
df['aws'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
```

```python
df['excel'] = df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
```

```python
df.head()
```

```python
df.drop(labels=['Unnamed: 0'],axis=1,inplace=True)
```
